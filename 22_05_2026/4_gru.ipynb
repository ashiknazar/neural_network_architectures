- A Gated Recurrent Unit (GRU) is a type of recurrent neural network (RNN) architecture designed to handle sequence data and solve the vanishing gradient problem seen in traditional RNNs.
- It was introduced as a simpler alternative to LSTM.

A normal RNN updates memory like this:

$$
h_t = \tanh(Wx_t + Uh_{t-1})
$$

### Problem

- Old information fades away over long sequences.
- Gradients vanish during backpropagation.
