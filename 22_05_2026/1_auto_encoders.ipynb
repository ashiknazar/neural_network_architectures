An autoencoder is a type of neural network that learns:

> “How can I compress data into a smaller representation and then reconstruct it back?”

- it is mainly used for 
  - dimensionality reduction
  - feature selection
  - noise removal
  - anomaly detection
  - data compression
  - image generation(vae)

Suppose you have an image:
- Input: 28×28 image = 784 pixels
- Instead of storing all 784 values,
- the network compresses it into maybe 32 numbers
- then reconstructs the image from those 32 values

The compressed representation is called the:

**Latent Space / Bottleneck**

___

#### Architecture of an Autoencoder

Input → Encoder → Latent Space → Decoder → Output

___

### 1. Basic Autoencoder

Simple compression + reconstruction.

- dimensionality reduction

### 2. Sparse Autoencoder

Helps learn meaningful features.
