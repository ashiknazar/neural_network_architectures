```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Sample data
text = "I love to play football. Football is my favorite sport. I play football every weekend."

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

# Create input sequences
input_sequences = []
for line in text.split('.'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i + 1]
        input_sequences.append(n_gram_sequence)

# Pad sequences
max_sequence_length = max(len(x) for x in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')

# Create predictors and labels
X, y = input_sequences[:, :-1], input_sequences[:, -1]
y = tf.keras.utils.to_categorical(y, num_classes=total_words)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(total_words, 50, input_length=max_sequence_length - 1))
model.add(tf.keras.layers.GRU(100))
model.add(tf.keras.layers.Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=100, verbose=1)
```

```python
def predict_next_word(text):
    # Convert the input text to a sequence
    token_list = tokenizer.texts_to_sequences([text])[0]
    token_list = pad_sequences([token_list], maxlen=max_sequence_length - 1, padding='pre')
    predicted = model.predict(token_list, verbose=0)
    predicted_word_index = np.argmax(predicted, axis=-1)
    
    # Return the predicted word
    return tokenizer.index_word[predicted_word_index[0]]

# Example usage
input_text = "I love to play"
predicted_word = predict_next_word(input_text)
print(f"Predicted next word: {predicted_word}")
```
