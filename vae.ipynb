### Variational Encoder Distribution

In a Variational Autoencoder (VAE), the encoder models the latent variable distribution as:

$$
q(z \mid x) = \mathcal{N}(\mu(x), \sigma^2(x))
$$

where:
- $ \mu(x) $ is the mean vector produced by the encoder network  
- $ \sigma^2(x) $ is the variance vector produced by the encoder network  
- $ z \sim q(z \mid x) $ is sampled from this Gaussian distribution conditioned on input $ x $

### Reparameterization Trick in VAE

In a Variational Autoencoder (VAE), the latent variable \( z \) is computed using the reparameterization trick:

$$
z = \mu + \sigma \cdot \epsilon
$$

where:

- $\mu$: Mean of the latent distribution output by the encoder network for input $x$
- $\sigma$: Standard deviation of the latent distribution, computed as $\sigma = \exp(0.5 \cdot \log \sigma^2)$
- $\epsilon$: Random noise sampled from a standard normal distribution $\mathcal{N}(0, I)$

### Key Idea
Instead of directly sampling $z \sim \mathcal{N}(\mu, \sigma^2)$, we rewrite it as a deterministic transformation of noise $\epsilon$, which allows gradients to flow through $\mu$ and $\sigma$ during training.

### Intuition
- $\mu$ shifts the center of the distribution  
- $\sigma$ controls the spread (uncertainty)  
- $\epsilon$ injects randomness in a differentiable way
